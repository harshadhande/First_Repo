#Importing Pandas

```python
import pandas as pd
import numpy as np
```

#Creating Series Objects from list

```python
#string
country=['India','Pakistan','Bangladesh','Nepal','Bhutan']

pd.Series(country)
```

```python
abc=[13,65,34,98,70,34]
pqr=pd.Series(abc)
print(pqr)
```

```python
#custom index
marks=[95,87,85,94,47]
sub=['DA','DMS','PT&RP','BHR','UHV']
pd.Series(marks,index=sub)
```

```python
#setting a name
pd.Series(marks,index=sub,name='marks of Harsha')
```

#Creating Series Object from Dictionary

```python
marks1={'DA':95,'DMS':87,'PT&RP':85,'BHR':94,'UHV':47}
abc=pd.Series(marks1,name='marks of Harsha')
abc
```

```python
marks1={'DA':95,'DMS':87,'PT&RP':85,'BHR':np.nan,'UHV':47}
xyz=pd.Series(marks1,name='marks of Harsha')
xyz
```
